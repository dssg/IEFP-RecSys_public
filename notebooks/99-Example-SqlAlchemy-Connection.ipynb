# SqlAlchemy

## Imports

```python
import os
import yaml
import pandas as pd
from urllib.parse import quote_plus  # PY2: from urllib import quote_plus
from sqlalchemy.engine import create_engine
from sqlalchemy.sql.expression import select
from sqlalchemy.sql.functions import func
from sqlalchemy.sql.schema import Table, MetaData
```

```python
aws_cred = yaml.load(open("../conf/local/credentials.yml"), Loader=yaml.FullLoader)
```

## SqlAlchemy connection

```python
# Basic connection

conn_str = 'awsathena+rest://{aws_access_key_id}:{aws_secret_access_key}@athena.{region_name}.amazonaws.com:443/'\
           '{schema_name}?s3_staging_dir={s3_staging_dir}'
engine = create_engine(conn_str.format(
    aws_access_key_id=quote_plus(aws_cred['dssg']['aws_access_key_id']),
    aws_secret_access_key=quote_plus(aws_cred['dssg']['aws_secret_access_key']),
    region_name='eu-west-2',
    schema_name='default',
    s3_staging_dir=quote_plus('s3://iefp-unemployment/SIGAE/')))

# SQL query

query = "select * from sigae.emp_ofertas LIMIT 5;"
df = pd.read_sql_query(query, engine)
```
